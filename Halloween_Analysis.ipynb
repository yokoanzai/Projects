```
plt.style.use("fivethirtyeight")

plt.plot(data["Year"], data["6pm"], label = "@6pm")
plt.plot(data["Year"], data["6:30pm"])
plt.plot(data["Year"], data["7pm"])
plt.plot(data["Year"], data["7:30pm"])
plt.plot(data["Year"], data["8pm"])
plt.plot(data["Year"], data["Total (8:15pm)"])

plt.ylabel('# Number of Kids')
plt.xlabel('Years')
plt.title('Halloween Data')

plt.axhline(y = 0, color = "black")
plt.axvline(x = 2008.1, color = "black", linewidth=1.3)

plt.plot(figsizes = (6,4))

plt.legend(loc = "upper right", ncol = 3, fontsize = 8)
```
